```python
# Target variable is species - Setosa, Virginica and Versicolor - Multiclass classification problem

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# Independent features

X = df.iloc[:,:-1]
X
```

```python
# Dependent Features

y = iris.target
y
```

```python
X,y
```

```python
### train,test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## before Postpruning

treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled=True)
```

### Tree Structure

Tree splitted based on Petal_Length <=0.8 and spliting criteria used here is "gini" - 0 to 0.5, 0 meaning pure split (leaf node) when 0.5 meaning impure split.

after 1st split, all 31 setosa went to left side of the root node which means setosa comes when petal length is less than or equal to 0.8..

Further split done based on petal length and sepal width - 34 Versicolor wend to left node of the decision node whereas 30 Virginica goes to right node.. max number of categories assigned at this stage, so we do not need further split..

In this case, we will do post prunning, remove further unnecessary splits.

## Postpruning

```python
treemodel1 = DecisionTreeClassifier(max_depth=2)
```

```python
treemodel1.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(treemodel1, filled=True)
```

## Prediction

```python
y_pred = treemodel1.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
